# 作業. 梯度下降法(1)

## 1. $x^3-2x+100$

```python
#確實import所需模組
import numpy as np
from matplotlib import pyplot as plt
from sympy import *
import math
```

```python
#當函數較複雜時，無法用手算一階導數時，可是使用此方法

x = Symbol('x')
y = x ** 3 - 2 * x + 100 
print(y)
yprime = y.diff(x)#y.diff(x)對函數做偏微分

# 目標函數 注意：要帶入實際值算出數值
def f(x_value): 
    return (x_value) ** 3 - 2 * (x_value) + 100 

# 目標函數一階導數:dy/dx 呼叫上面的yprime並將x_value取代x帶入後再轉為evalf()實數
df = lambda x_value: yprime.subs(x, x_value).evalf()

# 梯度下降函數
# x_start: x的起始點
# df: 目標函數的一階導數
# epochs: 執行週期
# lr: 學習率(梯度下降的速度)
def GD(w_start, epochs, lr):    
    w_list = np.zeros(epochs+1) #記錄下方圖中紅色的點   
    w = w_start    
    w_list[0] = w    
    for i in range(epochs):  #跑回圈找最小值       
        # w更新 w_new = w — learning_rate(學習率) * gradient(梯度)        
        w -= lr * df(w)         
        w_list[i+1] = w  #記錄每次新的w  
    return w_list
```

```python
# 超參數(Hyperparameters)
w_start = 4     # 起始權重
epochs = 1500  # 執行週期數 
lr = 0.01      # 學習率 

# 梯度下降法 
# *** Function 可以直接當參數傳遞 ***
w = GD(w_start, epochs, lr=lr) 
print('w的變化軌跡：', np.around(w, 2))

# 設定中文字型
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  
# 矯正負號
plt.rcParams['axes.unicode_minus'] = False

color = 'r'    
t = np.arange(-6.0, 6.0, 0.01)
plt.plot(t, [f(x) for x in t], c='b')
plt.plot(w, [f(x) for x in w], c=color, label='lr={}'.format(lr))    
plt.scatter(w, [f(x) for x in w], c=color, ) 

plt.title('梯度下降', fontsize=20)
plt.xlabel('w', fontsize=20)
plt.ylabel('損失函數', fontsize=20)
plt.show()
```
